## Categories Column EDA

From my EDA, I see that there are 194 unique categories. 194 categories is way too many categories to include in an analysis, so I also did a breakdown of how many times each categories appears in the dataset. I want to only use the most popular/relevant

```python
categories = []

stop_words = ['Food', 'Restaurants'] #filter out these words because they are redundant
for cat in data['categories']:
    cat_split = cat.split()
    nums_removed = [i for i in cat_split if i.isalpha()] # remove numbers and puncuation 
    stop_removed = [i for i in nums_removed if i not in stop_words] # remove the stop words
    for i in stop_removed:
        categories.append(i)
        
categories_unique = set(categories) # all the unique categories in the column
    
categories_unique
```

## Combine parking columns to reduce the number of columns in the dataset

```python
def free_parking(row):
    # if there's lot parking, then free
    # if there's valet parking that's validated, then free
    # if there's street and garage parking that's validated, then free
    # else not free
    if (row['Lot'] == True) or ((row['Valet'] == True) and (row['Validated'] == True)) or ((row['Street'] == True) and (row['Validated'] == True)) or ((row['Garage'] == True) and (row['Validated'] == True)):
        parking = 1
    else:
        parking = 0
        
    return parking
```

## Encode da variables

```python
def encode_prices(x):
    if x == '1':
        return 0
    if x ==  '2':
        return 1
    if x ==  '3':
        return 2
    if x ==  '4':
        return 3
    if pd.isna(x):
        return 0

def encode(x):
    if (x == False) or (x == None) or (x == 'False') or (x == 'None'):
        return 0
    if pd.isna(x):
        return 0
    if (x == True) or (x == 'True'):
        return 1
    
def encode_wifi(x):
    if (x == "'free'") or (x == "u'free'"):
        return 1
    else:
        return 0

def encode_states(x): # regions defined by census    
    if (x == 'AZ') or (x == 'CA') or (x == 'NV'): # west
        return 0 
    if (x == 'AB') or (x == 'DE') or (x == 'FL') or (x == 'LA') or (x == 'TN'): # south
        return 1
    if (x == 'ID') or (x == 'IL') or (x == 'IN') or (x == 'MO'): #midwest
        return 2
    if (x == 'NJ') or (x == 'PA'): # east
        return 3
```

```python
data_encoded = data.copy()
data_encoded['RestaurantsPriceRange'] = data_encoded['RestaurantsPriceRange'].apply(encode_prices)
```

```python
data_encoded['RestaurantsDelivery'] = data_encoded['RestaurantsDelivery'].apply(encode)
data_encoded['RestaurantsTakeOut'] = data_encoded['RestaurantsTakeOut'].apply(encode)
data_encoded['Hipster'] = data_encoded['Hipster'].apply(encode)
data_encoded['Touristy'] = data_encoded['Touristy'].apply(encode)
data_encoded['Trendy'] = data_encoded['Trendy'].apply(encode)
data_encoded['Upscale'] = data_encoded['Upscale'].apply(encode)
data_encoded['WiFi'] = data_encoded['WiFi'].apply(encode_wifi)
```

```python
data_encoded['Regions'] = data_encoded['state'].apply(encode_states)
```
