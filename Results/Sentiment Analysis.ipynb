```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
df = pd.read_csv('cleaned_dataframe.csv')
df.head()
```

```python
# REMOVE NaN VALUES AND EMPTY STRINGS:
df.dropna(inplace=True)

blanks = []  # start with an empty list

for i,row in df.iterrows():
    if type(row['text'])==str:            
        if row['text'].isspace():        
            blanks.append(row['Unnamed: 0'])     

df.drop(blanks, inplace=True)
```

```python
df
```

```python
df['scores'] = df['text'].apply(lambda review: sid.polarity_scores(review))

df.head()
```

```python
df['compound']  = df['scores'].apply(lambda score_dict: score_dict['compound'])

df.head()
```

```python
df['compound'].hist(bins = 8)
```

```python
def encode_score(x):
    if x <= -0.75: # very negative
        return 0
    if (x > -0.75) and (x <= -0.25): # negative
        return 1
    if (x > -0.25) and (x <= 0): # somewhat negative
        return 2
    if (x > 0) and (x < 0.25): # somewhat positive
        return 3
    if (x >= 0.25) and (x < 0.75): # positive
        return 4
    if (x >= 0.75): # very positive
        return 5
```

```python
df['comp_score'] = df['compound'].apply(encode_score)

df.head()
```

## Compound by restaurant

```python
df_mean = df.groupby(['business_id']).mean()
df_mean
```

```python
df_median = df.groupby(['business_id']).median()
df_median
```

```python
df_median = df_median.drop(columns = ['Unnamed: 0', 'index'])
df_mean = df_mean.drop(columns = ['Unnamed: 0', 'index'])
```
